```python
import numpy as np
```

```python
tol = 1.0e-8
```

```python
# Set up A
A = np.array([
[4, -1, 0, 0, 0],
[-1, 4, -1, 0, 0],
[0, -1, 4, -1, 0],
[0,0,-1,4,-1],
[0,0,0,-1,4],
])
A
```

```python
# set up d
d= np.array ([3,2,2,2,3], dtype=float)
d
```

```python
# initialize x
x = np.zeros(5)
x
```

```python
# define max number of sweeps
nsweep = 20
```

```python
for k in range(nsweep):
    x_old = x.copy()
    for i in range(5):
        sigma = 0.0
        for j in range(5):
            if j!=i:
                sigma = sigma + A[i,j]*x[j]
        x[i] = (d[i] - sigma)/A[i,i]
    residual = np.linalg.norm(x-x_old)
    if residual < tol:
        break


print(x)
print(residual)
        
```
