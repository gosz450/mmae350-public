## 3. NumPy Arrays

Most computational mechanics is vectorized: we work with arrays, not single numbers.

```python
import numpy as np

x = np.linspace(0.0, 1.0, 6)  # 6 points from 0 to 1
x
```

```python
dx = x[1] - x[0]
dx
```

### Array slicing

Slicing is used constantly when building numerical methods.

```python
print("x[2:5] =", x[2:5])   # entries 2, 3, 4
print("x[::2] =", x[::2])   # every other entry
```

## 4. Linear Algebra: Solve a Small System

Many discretizations lead to linear systems of the form:

\[
A\mathbf{x} = \mathbf{b}.
\]

We will solve a small example and interpret the result.

```python
A = np.array([[3.0, 1.0],
              [1.0, 2.0]])
b = np.array([9.0, 8.0])

sol = np.linalg.solve(A, b)
sol
```

## 5. Plotting (Matplotlib)

Plots are one of the main ways we interpret computational results.

```python
import matplotlib.pyplot as plt

x = np.linspace(0.0, 1.0, 100)  # 100 points from 0 to 1
y = np.sin(2*np.pi*x)

plt.figure()
plt.plot(x, y, marker="o")
plt.xlabel("x")
plt.ylabel("y")
plt.title("A simple plot")
plt.show()
```
