## Gauss--Seidel

```python
import numpy as np
```

```python
#Pseudo code
#
# set up the A matrix and rhs b vector
# determine the number of rows in matrix A (or b)
# set up initial guess X vector
# decide maximum number of iterations (niter)
# loop over number of iterations
    # X_new = X.copy()
    # for each row in A:
        # compute the first sum: sumkp1
        # compute the second sum: sumk
        # compute X_new
    # update X
    # X = X_new
```

```python
# set up A and b
A = np.array([[4,-1,0,0],[-1,4,-1,0],[0,-1,4,-1],[0,0,-1,3]])
b = np.array([[15],[10],[10],[10]])
print(A,b)
```

```python
# calculate number of rows in A
nrows = len(b)
nrows
```

```python
# set up initial guess vector X
X = np.zeros((nrows, 1))
X
```

```python
# GS Iterations loop
niter = 20
tol = 1.0e-12
for k in range(niter):
    for i in range(nrows):
        bi = b[i,0]
        aii = A[i,i]
        sumkp1 = 0.0
        for j in range(i):
            sumkp1 = sumkp1 + A[i,j]*X[j]
        sumk=0.0
        for j in range(i+1,nrows):
            sumk = sumk+ A[i,j]*X[j]
        X[i] = (bi-sumkp1 - sumk)/aii
        # compute residual
    R_rel = np.linalg.norm(b - A@X)/np.linalg.norm(b)
    if R_rel < tol:
        print(R_rel)
        print(tol)
        break
print(k)
```
