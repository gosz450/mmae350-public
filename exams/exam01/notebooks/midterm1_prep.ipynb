```python
# Set up numpy arrays A and x
A=np.array([[2,-1,-2],[-1,2,-1],[-2,0,2]])
x=np.array([1,2,3])
b = np.zeros(3)
# suppose A is of order mxm and x is mx1
m = len(A)
for i in range(m):
    b[i]=0
    for j in range(m):
        b[i] = b[i]+A[i,j]*x[j]
```
